```python
import numpy as np
import math
import pandas as pd

dt=pd.read_csv("/content/data.csv", encoding='latin1')
```

## step 1

```python
x=dt["no2"]
x.head()
r=102317119
a_r = 0.05 * (r % 7)
b_r = 0.3 * ((r % 5) + 1)

print("a_r =", a_r)
print("b_r =", b_r)

z=x+a_r*np.sin(b_r*x)
dt["z_transformed"] = z
dt[["no2","z_transformed"]].head()
```

## step 2

```python
z_vals=dt["z_transformed"].dropna().values
z_vals[:5]

mu=np.mean(z_vals)
print("mu:",mu)

variance=np.var(z_vals)
print("variance:",variance)

lam=1/(2* variance)
print("lambda:",lam)

c=np.sqrt(lam/np.pi)
print("c:",c)
```
